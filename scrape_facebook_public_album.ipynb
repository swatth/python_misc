## Use this notebook to scrape photos from facebook public album page

```python
import os
import csv
import time
import urllib.request
from lxml import html
from contextlib import closing
from selenium.webdriver import Chrome # pip install selenium
from selenium.webdriver.support.ui import WebDriverWait


def scrape_facebook_public_album(url, target_path, max_retry = 3):

    #attempt to open the page up to max_retry
    for i in range(max_retry):
        try:
            with closing(Chrome()) as browser:
                browser.get(url)
                #time.sleep(3)
                #driver.execute_script("window.scrollTo(0, Y);")
                time.sleep(30)
                
                # wait for the page to load
                #WebDriverWait(browser, timeout=30).until(
                    #lambda driver: driver.find_elements_by_xpath('//div[@testid="UFI2CommentLink"]'))
                # store it to string variable
                page_source = browser.page_source

        except Exception as e: 
            time.sleep(60*(i+1))
            print("attempt: " + str(i) + " " + str(e) + " url: " + url)
            continue
        #print("attempt: " + str(i) + " outside try")
        break

    #print(page_source)

    page = html.fromstring(page_source)

    #retrieved_links = page.xpath("//img")
    #print(retrieved_links)



    img_links = page.xpath('//*[@id="content_container"]//a')

    print(len(img_links))

    if len(img_links) > 0:

        if not os.path.exists(target_path):
            os.makedirs(target_path)

        for link in img_links:
            img_url = link.get("href")
            #print(img_url)
            
            #check to ensure the link is full link (starting with http...)
            if img_url.find("http") == 0:

                open_link_and_save_img(img_url, target_path)




    


def open_link_and_save_img(url, target_path, max_retry = 3):
    
    #attempt to open the page up to max_retry
    for i in range(max_retry):
        try:
            with closing(Chrome()) as browser:
                browser.get(url)
                time.sleep(10)
                # wait for the page to load
                #WebDriverWait(browser, timeout=30).until(
                #    lambda driver: driver.find_elements_by_xpath("//div[@class = 'uiContextualLayerPositioner uiLayer hidden_elem']"))
                # store it to string variable
                page_source = browser.page_source

        except Exception as e: 
            time.sleep(60*(i+1))
            print("attempt: " + str(i) + " " + str(e) + " url: " + url)
            continue
        #print("attempt: " + str(i) + " outside try")
        break

    #print(page_source)

    page = html.fromstring(page_source)

    #retrieved_links = page.xpath("//img")
    #print(retrieved_links)

    img_links = page.xpath('//a')

    print(len(img_links))

    if len(img_links) > 0:

        for link in img_links:
            img_url = link.get("data-ploi")

            #print(img_url)

            if img_url is not None and len(img_url) > 0:
                img_file = os.path.basename(img_url)
                print(img_url)
                img_file = img_file[0:img_file.find('.jpg?')] + '.jpg'
                
                print(img_file)

                save_file = os.path.join(target_path,img_file)

                #attempt to save the image up to max_retry
                for i in range(max_retry):
                    try:
                        urllib.request.urlretrieve(img_url,save_file)
                        time.sleep(2)
                    except Exception as e: 
                        time.sleep(10*(i+1))
                        print("attempt: " + str(i) + " " + str(e) + " url: " + img_url)
                        continue
                    #print("attempt: " + str(i) + " outside try")
                    break






def load_runner_list(filename):
    #format file: csv
    #column: gender-age-group, race no., name, country, finish time, pace, pace_group
    
    with open(filename, encoding="utf8") as f:
        reader = csv.reader(f)
        bib_numbers = []
        file_paths = []
        for row in reader:
            bib_number = row[1]
            file_path = os.path.join(row[0],row[6])
            
            bib_numbers.append(bib_number)
            file_paths.append(file_path)
            

    return [bib_numbers, file_paths]
```
